```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
```

```python
f = 1000 * np.array([0.05, 0.1, 0.200, 0.500, 1, 2, 5, 10, 20, 50, 100, 500])
U_out = np.array([2, 2, 2, 2, 2, 2, 1.85, 1.54, 1.05, 0.475, 0.245, 0.048])
U_a = 0.001 * np.array([5, 5.5, 11, 27, 54, 106, 250, 417, 560, 625, 590, 240])
A = 101 * U_out / U_a
print(A)
lg_f = np.log10(f)
lg_A_20 = 20 * np.log10(A)

fig, ax = plt.subplots() 
ax.grid()
ax.scatter(lg_f, lg_A_20, s=7)
plt.xlabel("lg(f)")
plt.ylabel("20 * lg(A), дБ")

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(lg_f[1:-2].reshape((lg_f[1:-2].size, 1)), lg_A_20[1:-2])
x0 = np.linspace(0, 7, 150).reshape((150, 1))
y0 = linear_regressor.predict(x0)
ax.plot(x0, y0)
print(linear_regressor.coef_)

lg_A0_20 = lg_A_20[0]
print(np.power(10, 3 / 20 + lg_f[1]))

fig.savefig('77_2')
linear_regressor.predict(np.array([6.59]).reshape(1, 1))
```

```python
f = 1000 * np.array([0.01, 0.02, 0.05, 0.200, 0.500, 1, 2, 5, 10, 20, 50, 100])
U_out = 10 * np.array([0.5, 0.5, 0.5, 0.51, 0.51, 0.51, 0.51, 0.51, 0.505, 0.467, 0.318, 0.187])
U_in = 0.05 * np.ones(12)
A = U_out / U_in

fig, ax = plt.subplots() 
ax.grid()
ax.scatter(f, A, s=7)
plt.xlabel("f, Hz")
plt.ylabel("A")

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(f[7:-1].reshape((f[7:-1].size, 1)), A[7:-1])
x0 = np.linspace(0, 100000, 150).reshape((150, 1))
y0 = linear_regressor.predict(x0)
# ax.plot(x0, y0)
print(linear_regressor.coef_)

fig.savefig('77_3')
linear_regressor.predict(np.array([43000]).reshape(1, 1))
```
