```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
```

```python
W = np.array([0.86, 0.934, 1.39, 1.69, 2.36, 2.66, 3.96, 4.17, 6.16, 7.72]) 
T = np.array([1200, 1320, 1440, 1500, 1600, 1690, 1810, 1940, 2050, 2150])

fig, ax = plt.subplots() 
ax.grid()
ax.scatter(T, W, s=5)
plt.xlabel("T, K")
plt.ylabel("W, Watt")
fig.savefig('W(T)')
```

```python
fig, ax = plt.subplots() 
ax.grid()
ax.scatter(np.log(T), np.log(W), s=5)
plt.xlabel("ln(T / K)")
plt.ylabel("ln(W / Watt)")


linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(np.log(T).reshape((np.log(T).size, 1)), np.log(W))


x0 = np.linspace(7, 7.8, 150).reshape((150, 1))
y0 = linear_regressor.predict(x0)

ax.plot(x0, y0)
print(linear_regressor.coef_)

fig.savefig('lnW(lnT)')
```
