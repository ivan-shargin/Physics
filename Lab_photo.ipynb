```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
```

ВАХ для минимальной длины волны.

```python

U_I = np.array([0, 1, 2, 4, 6, 8, 10, 20, 40, 70, 103, 140, 165, 200, 300, 505, 505, 505, 507]) 
U_e = np.array([-601, -595, -590, -576, -562, -545, -529, -467, -373, -267, -169, -63, 0, 90, 34, 727, 650, 600, 500])

fig, ax = plt.subplots() 
ax.grid()
ax.scatter(U_e, U_I, s=5)
plt.xlabel("U_electrode")
plt.ylabel("U_I")
```

ВАХ для максимальной длины волны

```python
U_I_1 = np.array([0, 3, 5, 10, 15, 20, 25, 30, 35, 40, 50]) 
U_e_1 = np.array([-124, -115, -107, -91, -78, -64, -53, -42, -31, -22, 0])

fig, ax = plt.subplots() 
ax.grid()
ax.scatter(U_e_1, U_I_1, s=5)
plt.xlabel("U_electrode")
plt.ylabel("U_I")
```

1я линия спектра

```python
Sqrt_U_I_1 = np.sqrt(U_I_1)
print(Sqrt_U_I_1)

fig, ax = plt.subplots() 
ax.grid()
ax.scatter(U_e_1, Sqrt_U_I_1, s=5)
plt.xlabel("U_electrode")
plt.ylabel("U_I ^ 0.5")

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(Sqrt_U_I_1.reshape((Sqrt_U_I_1.size, 1)), U_e_1)


x0 = np.linspace(-1, 8, 150).reshape((150, 1))
y0 = linear_regressor.predict(x0)

ax.plot(y0, x0)

print(linear_regressor.predict(np.array([0]).reshape((1, 1))))
```
