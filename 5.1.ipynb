```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
```

```python
l = np.array([0, 0.48, 0.93, 1.33, 1.80, 2.28, 2.78])
N_plus_phon = np.array([1650973, 858713, 405663, 193967, 105621, 58347, 31335])
N = N_plus_phon - 1539
ln_N = np.log(N)
delta_lnN = 1 / np.sqrt(N)
delta_l = np.linspace(0, 6 * 0.005, 7)

fig, ax = plt.subplots() 
ax.grid()
ax.scatter(l, ln_N, s=3)
ax.errorbar(l, ln_N, yerr=delta_lnN, xerr=delta_l, linestyle = '')
plt.xlabel("l, sm")
plt.ylabel("ln(N)")

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(l.reshape((l.size, 1)), ln_N)
x0 = np.linspace(0, 3, 150).reshape((150, 1))
y0 = linear_regressor.predict(x0)
ax.plot(x0, y0)
print(linear_regressor.coef_)


fig.savefig('5_1_1')
```
