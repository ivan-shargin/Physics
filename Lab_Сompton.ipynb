```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
```

```python
angles = np.linspace(0, math.pi / 2, 10)
Ns = np.array([388, 408, 383, 341, 321, 321, 303, 281, 259, 242]) 
x = (1 - np.cos(angles))
y = 1 / Ns

fig, ax = plt.subplots() 
ax.scatter(x, y, s=5)

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(x.reshape((10, 1)), y,)
print(linear_regressor.coef_)

x0 = np.linspace(0, 1, 1000).reshape((1000, 1))
y0 = linear_regressor.predict(x0)

ax.plot(x0, y0)
ax.grid()

fig.savefig('gg')
```
