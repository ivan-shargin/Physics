```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
```

```python
x = 0.01 * np.array([12, 13, 10, 7, 4, 9, 6, 11, 5]) # 5, 7, 4,  10, 6 - плохие точки
T = 0.001 * np.array([941, 1000, 918, 840, 853, 869, 919, 897, 836]) # 836, 840, 853, 918, 919 - плохие точки
T_2 = T * T
beta = 0.35 / 2 - x
alpha = (0.35 * 0.35 / 12 + beta * beta) / beta
fig, ax = plt.subplots() 
ax.grid()
ax.scatter( alpha, T_2, s=25)
plt.xlabel("alpha,   cm")
plt.ylabel("T^2,   c^2")

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(alpha.reshape((alpha.size, 1)), T_2)


x0 = np.linspace(0.19, 0.28, 150).reshape((150, 1))
y0 = linear_regressor.predict(x0)

ax.plot(x0, y0)

x_1 = linear_regressor.predict(np.array([0]).reshape((1, 1)))[0]
print(4 * math.pi * math.pi / linear_regressor.coef_)
print(np.sqrt(x_1))

fig.savefig('Ruslan')
```
