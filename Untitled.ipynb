```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
```

```python
N = np.array([1694, 1494, 1623, 846])
E = np.array([1.332, 1.173, 1.274, 0.6617])


fig, ax = plt.subplots() 
ax.grid()
ax.scatter(N, E, s=12)
plt.xlabel("N")
plt.ylabel("E, Mev")

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(N.reshape((N.size, 1)), E)
x0 = np.linspace(800, 1700, 150).reshape((150, 1))
y0 = linear_regressor.predict(x0)
ax.plot(x0, y0)
print(linear_regressor.coef_)
k = linear_regressor.coef_

print(linear_regressor.predict(np.array([109]).reshape(-1, 1)))

fig.savefig('calibr')
```

```python
Exp = np.array([0.919, 1.04, 0.465, 0.031, 0.246])
teor = np.array([1.117, 1.061, 0.477, 0.013, 0.199])

fig, ax = plt.subplots() 
ax.grid()
ax.scatter(Exp, teor, s=5)
plt.xlabel("Exp, Mev")
plt.ylabel("teor, Mev")

x0 = np.linspace(0, 1.2, 150)
plt.plot(x0, x0)

fig.savefig('comp_edge')
```

```python
Ei = np.array([1.332, 1.173, 1.274, 0.6617, 0.346])
Ri = np.array([0.045, 0.049, 0.048, 0.062, 0.092])
Ei_inv = 1 / Ei
Ri_2 = Ri * Ri
fig, ax = plt.subplots() 
ax.grid()
ax.scatter(Ei_inv, Ri_2, s=10)
plt.xlabel("1/Ei, 1/Mev")
plt.ylabel("Ri^2")

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(Ei_inv.reshape((Ei_inv.size, 1)), Ri_2)
x0 = np.linspace(0, 3, 150).reshape((150, 1))
y0 = linear_regressor.predict(x0)
ax.plot(x0, y0)
print(linear_regressor.coef_)
k = linear_regressor.coef_

fig.savefig('Ri_2')
```
