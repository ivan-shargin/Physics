```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
```

```python
d = np.linspace(0, 1.95 * 9, 10) 
N = np.array([231, 209, 171, 193, 180, 164, 165, 166, 193, 158])
delta_N = np.round(np.sqrt(N), decimals=-1)
delta_d = 0.05 / 1.95 * d
print(delta_d)
fig, ax = plt.subplots() 
ax.grid()
ax.errorbar(d, N, yerr=delta_N, xerr=delta_d, linestyle=' ')
plt.xlabel("d")
plt.ylabel("N")
fig.savefig('Cosmos')
```

```python
fig, ax = plt.subplots() 
ax.grid()
ax.scatter(d, np.log(N))
plt.xlabel("d")
plt.ylabel("log(N)")
fig.savefig('Cosmos')

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(d.reshape((d.size, 1)), np.log(N))

x0 = np.linspace(0, 20, 100).reshape((100, 1))
y0 = linear_regressor.predict(x0)

ax.plot(x0, y0)
print(linear_regressor.coef_[0])
```
